```python
from transformers import pipeline
classifier = pipeline('sentiment-analysis')
classifier('I am reading the greatest NLP book ever!')
```

```python
## truncate Body length to 512
def shorten(sentence):
  if isinstance(sentence, str):
    if len(sentence) > 512:
      return sentence[:512]
  return sentence
```

```python
df["Body"].describe()
```

```python
df["Body"] = df["Body"].apply(shorten)
```

```python
def predict_sentiment(sentence):
  pred = classifier(sentence)[0]
  return pred["label"], pred["score"]
```

```python
predict_sentiment(df["Body"][0])
```

```python
df["Sentiment and score"] = df.loc[:, "Body"].apply(predict_sentiment)
```

```python
def sort_sentiment(prediction):
  return prediction[0]

def sort_score(prediction):
  return prediction[1]
```

```python
df["Sentiment"] = df["Sentiment and score"].apply(sort_sentiment)
df["Score"] = df["Sentiment and score"].apply(sort_score)
```
