```python
sample.columns.tolist()[1:]
```

```python
sample = sample.reindex(sample.columns.tolist()[1:] + ['ScoreMixed','ScoreNegative','ScoreNeutral','ScorePositive','Sentiment'], axis=1) 
```

```python
sample.head
```

```python
sample.columns
```

```python
sample=sample.replace(np.nan, '', regex=True)
```

```python
sample[sample["Sentiment"] == ''].index
```

```python
for i in sample[sample["Sentiment"] == ''].index:
    print ( sample.at[i,"Body"][:10])
    text = sample.at[i,"Body"]
    resp=comprehend.detect_sentiment(Text=text[:4800], LanguageCode='en')
    sample.at[i,"Sentiment"] = resp["Sentiment"]
    sample.at[i,'ScoreMixed'] = resp["SentimentScore"]["Mixed"]
    sample.at[i,'ScoreNegative'] = resp["SentimentScore"]["Negative"]
    sample.at[i,'ScoreNeutral'] =  resp["SentimentScore"]["Neutral"]
    sample.at[i,'ScorePositive'] = resp["SentimentScore"]["Positive"]
    
```
